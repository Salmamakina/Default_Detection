```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('Picture1.jpg')
# Display the image
plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis("off");
```

```python
def est_noir_et_blanc(image):
    # Convert the image to gray-scale
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculer la différence absolue entre l'image en couleur et l'image en niveaux de gris
    diff = cv2.absdiff(image, cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR))

    # Vérifier si tous les canaux de couleurs sont proches de zéro (noir)
    is_black_and_white = (diff.sum() == 0)

    return is_black_and_white
```

```python
# Make a copy of the source image.
image_copy = image.copy()
if est_noir_et_blanc(image_copy) :
  Gray_image = cv2.cvtColor(image_copy,cv2.COLOR_BGR2GRAY)
  contours, hierarchy = cv2.findContours(Gray_image, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
else:
  # Convert the image to gray-scale
  gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
  # Invert the colors
  gray_inverted = cv2.bitwise_not(gray)
  # create a binary thresholded image
  _, binary = cv2.threshold(gray_inverted, 120, 255, cv2.THRESH_BINARY)
  plt.imshow(binary, cmap="gray");plt.title("Binary Image");plt.axis("off");
  # Find all contours in the image
  contours, hierarchy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
#draw all contours
image_copy = cv2.drawContours(image_copy, contours, -1, (0, 0, 255), 1)

# Plot the image with contours
plt.figure(figsize=[15,15])
plt.subplot(122);plt.imshow(image_copy[:,:,::-1]);plt.title("With Contours");plt.axis('off');
l=[]

# Loop over the contours
for i,cont in enumerate(contours):
        # #Draw the ith contour
        # image_copy = cv2.drawContours(image.copy(), cont, -1, (255,0,0), 2)

        # #Add a subplot to the figure
        # plt.subplot(2, 2, i+1)

        # #Turn off the axis
        # plt.axis("off");plt.title('contour ' +str(i))

        # #Display the image in the subplot
        # plt.imshow(image_copy)
        # ##the area of all contours
        contour_area = cv2.contourArea(contours[i])
        print("the area of the contour number",i,"est ",contour_area)
        l.append(contour_area)
# Print the number of Contours returned
print("Number of Contours Returned: {}".format(len(contours)))
```

```python
# Charger les deux images
image1 = cv2.imread('Picture1.jpg', 0)
image2 = cv2.imread('output_image.jpg', 0)
# Calculer la différence absolue entre les deux images
diff_image = cv2.absdiff(image1, image2)

# Seuiller l'image de différence pour obtenir une image binaire
threshold_value = 68 # Ajustez ce seuil en fonction de vos images
_, threshold_image = cv2.threshold(diff_image, threshold_value, 255, cv2.THRESH_BINARY)

# Trouver les contours dans l'image binaire
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dessiner un cercle exactement autour de chaque contour de déformation
image_with_circles = cv2.cvtColor(image2, cv2.COLOR_GRAY2BGR)
for contour in contours:
    # Créer un masque binaire autour du contour actuel
    mask = np.zeros_like(diff_image)
    cv2.drawContours(mask, [contour], 0, 255, thickness=cv2.FILLED)

    # Trouver les coordonnées du rectangle de contour englobant
    x, y, w, h = cv2.boundingRect(contour)

    # Trouver les coordonnées du cercle englobant exactement le contour
    center, radius = cv2.minEnclosingCircle(contour)

    # Dessiner le cercle sur l'image originale
    cv2.circle(image_with_circles, (int(center[0]), int(center[1])), int(radius), (0, 0, 255), 2)

# Afficher l'image avec les cercles autour des régions de déformation
plt.imshow(image_with_circles)
```

```python
image1 = cv2.imread('output_image.jpg', 0)
image2 = cv2.imread('rect1.png', 0)
# Calculer la différence absolue entre les deux images
diff_image = cv2.absdiff(image1, image2)

# Seuiller l'image de différence pour obtenir une image binaire
threshold_value = 30  # Ajustez ce seuil en fonction de vos images
_, threshold_image = cv2.threshold(diff_image, threshold_value, 255, cv2.THRESH_BINARY)

# Trouver les contours dans l'image binaire
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calculer la superficie totale de l'objet
total_area = threshold_image.shape[0] * threshold_image.shape[1]

# Initialiser la superficie de la zone défectueuse
defective_area = 0

# Calculer la superficie de la zone défectueuse
for contour in contours:
    defective_area += cv2.contourArea(contour)

# Calculer le pourcentage de la zone défectueuse par rapport à la superficie totale
defect_percentage = (defective_area / total_area) * 100

print(f"Pourcentage de la zone défectueuse : {defect_percentage:.2f}%")
```
